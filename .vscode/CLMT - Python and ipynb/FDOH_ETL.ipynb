```python
#TB Code
start=time()
r=requests.get(url)
x=r.json()

tb=pd.DataFrame()
for i in range(len(x['features'])):
    tb=tb.append(pd.DataFrame(x['features'][i]['attributes'], index=['OBJECTID']))
tb['Cases']=1
end=time()
runtime=(end-start)/60
print(runtime)
tb.head()
```
